```python
#Read data from Wikipedia into a Pandas DataFrame
toronto_data=pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&direction=prev&oldid=926287641')
toronto_data=toronto_data[0]
#Replace "Not assigned" Boroughs with NaN
toronto_data['Borough'].replace('Not assigned', np.nan, inplace=True)
#Drop NaN Values
toronto_data.dropna(subset=['Borough'], axis=0, inplace=True)
#Reset Index
toronto_data.reset_index()
```

```python
!conda install -c conda-forge geocoder --yes
import geocoder # import geocoder
```

```python
# initialize your variable to None
lat_lng_coords = None

postal_code=toronto_data['Postcode'].unique()
# loop until you get the coordinates
while(lat_lng_coords is None):
  g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
  lat_lng_coords = g.latlng

latitude = lat_lng_coords[0]
longitude = lat_lng_coords[1]
```

```python
lat_lon=pd.read_csv('http://cocl.us/Geospatial_data')
toronto_data=toronto_data.merge(lat_lon, how='left', left_on='Postcode', right_on='Postal Code').drop(['Postal Code'], axis=1)
```

I decided to filter just the Boroughs that contains "Toronto" in their names

```python
real_toronto_data=toronto_data[toronto_data['Borough'].str.contains('Toronto')]
real_toronto_data
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[43.685988, -79.414312], zoom_start=12)

# add markers to map
for lat, lng, label in zip(real_toronto_data['Latitude'], real_toronto_data['Longitude'], real_toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
